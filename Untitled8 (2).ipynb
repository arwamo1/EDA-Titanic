```python
import pandas as pd

df = pd.read_csv('Titanic-Dataset.csv')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['Cabin'])
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.show()
```

```python
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival by Embarked Port')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Fare', data=df)
plt.title('Average Fare by Class')
plt.show()
```

Data Cleaning

We cleaned the dataset by handling missing values and removing unnecessary columns. The "Cabin" column was dropped because it contained too many missing values. Missing values in the "Age" column were filled using the median, and missing values in the "Embarked" column were filled using the most frequent value. These steps improved the quality of the dataset and made it suitable for analysis.

Visualization and Insights

The visualizations show several important patterns in the dataset. The survival count indicates that more passengers did not survive than those who survived. The gender-based plot shows that females had a higher survival rate compared to males. The passenger class analysis reveals that passengers in first class had a higher chance of survival. The age distribution shows that most passengers were between 20 and 40 years old. The boxplot indicates differences in age distribution between survivors and non-survivors. The embarked port analysis shows variations in survival based on boarding location. Finally, the fare analysis shows that higher-class passengers paid more on average.
